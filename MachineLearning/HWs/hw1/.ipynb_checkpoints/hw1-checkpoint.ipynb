## 训练

```python
dimension = 18 * 9 + 1
w = np.zeros((dimension, 1))
x_tra = np.concatenate([np.ones((4521, 1)), x_tra], axis=1)
learning_rate = 0.001
```

```python
iter_time = 5000
```

```python
x_tra.shape
```

```python
for t in range(iter_time):
    y_predict = np.dot(x_tra, w)
    diff = y_predict - y_tra
    loss = np.sum(np.power(diff, 2)) / 4521
    if t % 500 == 0:
        print("loss is", loss)
    gradient = np.dot(x_tra.T, diff) * 2 / 4521
    w -= learning_rate * gradient
```

## Adagrad

```python
x_tra = normal_x_train[:math.floor(len(x_train) * 0.8), :]
y_tra = y_train[:math.floor(len(x_train) * 0.8)]
x_val = normal_x_train[math.floor(len(x_train) * 0.8):, :]
y_val = y_train[math.floor(len(x_train) * 0.8):]
```

```python
dimension = 18 * 9 + 1
w = np.zeros((dimension, 1))
x_tra = np.concatenate([np.ones((4521, 1)), x_tra], axis=1)
learning_rate = 100
adagrad = np.zeros([dimension, 1])
eps = 0.0000000001
```

```python
iter_time = 10000
```

```python
for t in range(iter_time):
    y_predict = np.dot(x_tra, w)
    diff = y_predict - y_tra
    loss = np.sum(np.power(diff, 2)) / 4521
    if t % 500 == 0:
        print("loss is", loss)
    gradient = np.dot(x_tra.T, diff) * 2 / 4521
    adagrad += gradient ** 2
    w -= learning_rate * gradient / np.sqrt(adagrad + eps)
```
